```python
while(True):
    try:
        x = float(input("Introduce un numero"))
        y = 5
        print("La multiplicacion de" , y, "por", x, "es =", x*y , "\n")
    except:
        print("Ah ocurrido un error")
    else:
        print("Todo funciono bien")
        break
    finally:
        print("Fin del programa")
```

```python
from logging import exception
while(True):
    try:
        x = float(input("Introduce un numero"))
        y = 100
        print("La division de" , y, "por", x, "es =", y/x , "\n") # Changed x*y to y/x for division
    except ValueError:
        print("No es numerico el valor proporcionado")
    except Exception as e:
        print("Error", type(e).__name__)
    else:
        print("Todo funciono bien")
        break
    finally:
        print("Fin del programa")
```
